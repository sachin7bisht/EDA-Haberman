Info about Haberman Cancer

Information: The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.

Number of Instances: 306

Number of Attributes: 4 (including the class attribute)

Attribute Information:

1.Age of patient at time of operation (numerical)
2.Patient's year of operation (year - 1900, numerical)
3.Number of positive axillary nodes detected (numerical)
4.Survival status (class attribute)
  1 = the patient survived 5 years or longer
  2 = the patient died within 5 year

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


breast_cancer=pd.read_csv("/content/haberman_cancer.csv")
breast_cancer.columns=['Age','Op_year','axil_nodes','Surv_status']

print(breast_cancer.shape)
```

Histogram and CDF

```
# 1D scatter plot Survival Status
import numpy as np
surv_greater_5= breast_cancer.loc[breast_cancer['Surv_status'] == "Live"];
surv_lesser_5= breast_cancer.loc[breast_cancer['Surv_status'] == "Dead"];
plt.plot(surv_greater_5['axil_nodes'], np.zeros_like(surv_greater_5['axil_nodes']), 'X',label='Live')
plt.plot(surv_lesser_5['axil_nodes'], np.zeros_like(surv_lesser_5['axil_nodes']), 'o',label='Dead')
plt.title('1D scatter plot of nodes')
plt.legend()
plt.show()
```

observation:
Survial status of both live and dead are densely overlaping till 25 nodes 

```
# PDF of Age
sns.FacetGrid(breast_cancer,hue='Surv_status',size=5).map(  sns.distplot,'Age').add_legend()
plt.title(' PDF Distrubution plot of AGE')
plt.show()
```

Observation:
from the PDF plot we can understand people from age 20-40 have better chance of survival

```
# PDF of Operation Year
sns.FacetGrid(breast_cancer,hue='Surv_status',size=5).map(  sns.distplot,'Op_year').add_legend()
plt.title(' PDF Distrubution plot of Operation Year')
plt.show()
```

Observation:
we can observe from the PDF that operation taken between the years 1958-1962 have most survived patients

```
# PDF of Axil Nodes
sns.FacetGrid(breast_cancer,hue='Surv_status',size=5).map(  sns.distplot,'axil_nodes').add_legend()
plt.title('PDF Distribution plot of Axillary Nodes')
plt.show()
```

Observation:
Survival status of patients having axillary nodes between 0-5 have high chance of survival

```
# pdf and cdf of age
counts,bin_edges=np.histogram(breast_cancer['Age'],bins=10,density='True')
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='PDF')
plt.plot(bin_edges[1:],cdf,label='CDF')
plt.title('CDF and PDF of Age')
plt.legend()
plt.show()
```

```
# pdf and cdf of axialluery nodes
counts,bin_edges=np.histogram(breast_cancer['axil_nodes'],bins=10,density='True')
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='PDF')
plt.plot(bin_edges[1:],cdf,label='CDf')
plt.title('CDF and PDF of Auxillary Node')
plt.legend()
plt.show()
```

Obseravtion:
1. There are less than 10% patients who have nodes less than 10
2. 10-80% of patients have nodes between 0-5

Median, Percentile, Quantile, IQR, MAD


```
print('Mean: Survival greater than 5 year ')
print('Age',np.mean(surv_greater_5['Age']))
print('Axial Nodes',np.mean(surv_greater_5['axil_nodes']))
print('Mean:Survival Lesser than 5 year') 
print('Age',np.mean(surv_lesser_5['Age']))
print('Axial Nodes',np.mean(surv_lesser_5['axil_nodes']))
# mean with an outliner
print(np.mean(np.append(surv_greater_5["Age"],50)))
print(np.mean(np.append(surv_greater_5["axil_nodes"],50)))
# stn deviation
print(np.std(surv_greater_5["Age"]))
print(np.std(surv_lesser_5["Age"]))
print(np.std(surv_greater_5["axil_nodes"]))
print(np.std(surv_lesser_5["axil_nodes"]))
```

Observation:
1. Patients who have survived have node less than 3
2. Patients who have not survived have nodes approx. equal to 7

```
print('Median:Survival greater than 5 year ')
print('Age',np.median(surv_greater_5['Age']))
print('Axial Nodes',np.median(surv_greater_5['axil_nodes']))
print('Median:Survival Lesser than 5 year') 
print('Age',np.median(surv_lesser_5['Age']))
print('Axial Nodes',np.median(surv_lesser_5['axil_nodes']))
# mean with an outliner
print(np.median(np.append(surv_greater_5["Age"],50)))
print(np.median(np.append(surv_greater_5["axil_nodes"],50)))
```

Observation:

For survival Patients:
Median Age is 52 and Nodes are 0

For patients who survived less than 5 years
Median Age is 53 and Nodes are 4

```
print("Quantiles:Age")
print(np.percentile(surv_greater_5["Age"],np.arange(0, 100, 25)))
print(np.percentile(surv_lesser_5["Age"],np.arange(0, 100, 25)))
print("Quantiles:Axial Nodes")
print(np.percentile(surv_greater_5["axil_nodes"],np.arange(0, 100, 25)))
print(np.percentile(surv_lesser_5["axil_nodes"],np.arange(0, 100, 25)))


print("Quantiles:Age")
print(np.percentile(surv_greater_5["Age"],90))
print(np.percentile(surv_lesser_5["axil_nodes"],90))
print("Quantiles:Axial Nodes")
print(np.percentile(surv_greater_5["Age"],90))
print(np.percentile(surv_lesser_5["axil_nodes"],90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(surv_greater_5["axil_nodes"]))
print(robust.mad(surv_lesser_5["axil_nodes"]))
```
